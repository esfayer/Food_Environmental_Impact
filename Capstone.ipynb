```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
comp = pd.read_csv('../../Capstone/Food_Production.csv')
```

```python
comp.head()
```

```python
## filtering to only the columns I want represented
## shortening column names for easier use / understanding

pared = comp [['Food product','Greenhouse gas emissions per 1000kcal (kgCO₂eq per 1000kcal)',
               'Eutrophying emissions per 1000kcal (gPO₄eq per 1000kcal)',
                'Land use per 1000kcal (square miles per 1000kcal)',
               'Scarcity-weighted water use per 1000kcal (liters per 1000 kilocalories)']]

pared_short = pared.rename(columns={'Greenhouse gas emissions per 1000kcal (kgCO₂eq per 1000kcal)': 'GHG_Emissions',
               'Eutrophying emissions per 1000kcal (gPO₄eq per 1000kcal)': 'Eutrophying_Emissions',
                'Land use per 1000kcal (square miles per 1000kcal)':'Land_Use',
               'Scarcity-weighted water use per 1000kcal (liters per 1000 kilocalories)':'Scarcity_weighted_Water_Use'})

pared_short
```

```python
## dropping nas

pared_nona = pared_short.dropna(subset = ['Scarcity_weighted_Water_Use'])
pared_nona
```

```python
## resetting index now that n/a rows eliminated
new_df = pared_nona.reset_index(drop=True)
new_df
```

```python
## want to add columns for quartiles
new_df.describe()
```

```python
##adding GHG percentiles
new_df['GHG_Percentiles'] = new_df.GHG_Emissions.rank(pct = True)
new_df
```

```python
##adding the rest of the percentiles
new_df['Eutrophying_Percentiles'] = new_df.Eutrophying_Emissions.rank(pct = True)
new_df['Land_Percentiles'] = new_df.Land_Use.rank(pct = True)
new_df['Weighted_Water_Percentiles'] = new_df.Scarcity_weighted_Water_Use.rank(pct = True)
new_df
```

```python
new_df['enviro_sum_percentiles'] = (new_df['Eutrophying_Percentiles'] + new_df['Land_Percentiles'] + 
                          new_df['Weighted_Water_Percentiles'] + new_df['GHG_Percentiles'])
new_df
```

```python
new_df['enviro_sum'] = (new_df['GHG_Emissions'] + new_df['Land_Use'] + 
                          new_df['Scarcity_weighted_Water_Use'] + new_df['Eutrophying_Emissions'])
new_df
```

```python
df_melt_long=new_df.melt(id_vars='Food product',var_name='Measure Type',value_name='Value')
```
